# Histogram of `SalePrice`

The histograms will reveal the distributions of numerical variables, which will help decide the transformation to use in further analysis.  

```python
%matplotlib inline
plt.figure(figsize=(6, 4))
sns.histplot(df['SalePrice'], kde=True)
plt.title(f"Distribution of Sale Price")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

# Distribution of Variables

Below plots show the distribution of variables. 
Histograms and box plots are provided for numerical columns and 
Box plots are provided for categorical ones. 

```python
%matplotlib inline
numeric_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

for col in df:
    if col in numeric_cols:
        print(f"{col} is a numeric variable")
        fig, axs = plt.subplots(2, 1, figsize=(6, 6), gridspec_kw={'height_ratios': [1, 3]})

        # Boxplot (top)
        sns.boxplot(x=df[col].dropna(), ax=axs[0])
        axs[0].set(title=f"Boxplot: {col}")
        axs[0].set(xlabel='')
        axs[0].tick_params(axis='x', labelbottom=False)

        # Histogram with KDE (bottom)
        sns.histplot(df[col].dropna(), kde=True, ax=axs[1])
        axs[1].set(title=f"Histogram: {col}", xlabel=col, ylabel="Count")

        plt.tight_layout()
        plt.show()
    
    elif col in cat_cols:
        print(f"{col} is a categorical variable")
        plt.figure(figsize=(8, 4))
        df[col].value_counts().plot(kind='bar')
        plt.title(f"Category Counts: {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()
```
